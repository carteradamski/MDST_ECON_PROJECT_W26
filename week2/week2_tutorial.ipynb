# Customizing Plots with Matplotlib: A Comprehensive Guide

This tutorial will teach you how to customize various features of plots including colors, sizes, spacing, markers, styles, and more using matplotlib and pandas.

## Prerequisites

First, let's install and import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

## 1. Creating Sample Data

For the Tutorial, I will just be doing a simple Dataset (Generated by AI) and then you will all do similar stuff in the workbook with our actual data.

```python
# Create sample sales data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [15000, 18000, 22000, 25000, 28000, 32000, 30000, 35000, 33000, 38000, 42000, 45000],
    'Expenses': [12000, 13000, 15000, 16000, 18000, 20000, 19000, 21000, 20000, 22000, 24000, 26000],
    'Profit': [3000, 5000, 7000, 9000, 10000, 12000, 11000, 14000, 13000, 16000, 18000, 19000]
}

df = pd.DataFrame(data)
df.to_csv('sample_data.csv', index=False)
print("Sample CSV created successfully!")
print(df.head())
```

## 2. Reading CSV Files

The most basic operation is reading a CSV file into a pandas DataFrame:

```python
# Read the CSV file
df = pd.read_csv('sample_data.csv')

# Display first few rows
print("First 5 rows:")
print(df.head())
```

## 3. Line Plot Customization Basics

Let's explore how to customize line plots with different colors, markers, and styles:

```python
# Create a basic line plot with customizations
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], 
         marker='o',           # Marker style: 'o', 's', '^', 'D', '*', 'v', etc.
         linewidth=2.5,        # Line thickness
         markersize=8,         # Marker size
        color="#862301",      # Custom hex color (Sea Green)
         linestyle='-',        # Line style: '-', '--', '-.', ':'
         alpha=1,            # Transparency (0-1)
         label='Sales')

plt.xlabel('Month', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Amount ($)', fontsize=14, fontweight='bold', color='darkblue')
plt.title('Monthly Sales', fontsize=16, fontweight='bold', pad=20)
plt.legend(fontsize=12, framealpha=0.9, shadow=True)
plt.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

## 4. Color Variations and Palettes

Explore different ways to specify colors and create appealing color combinations:

```python
plt.figure(figsize=(12, 6))

# Different ways to specify colors:
# 1. Named colors: 'red', 'blue', 'green', etc.
# 2. Hex colors: '#FF5733', '#3498DB', etc.
# 3. RGB tuples: (0.2, 0.4, 0.8)
# 4. Built-in matplotlib colors: 'C0', 'C1', 'C2', etc.

plt.plot(df['Month'], df['Sales'], 
         marker='o', linewidth=3, markersize=10, 
         color='#1f77b4',      # Hex color
         label='Sales', markerfacecolor='lightblue', markeredgecolor='darkblue', markeredgewidth=2)

plt.plot(df['Month'], df['Expenses'], 
         marker='s', linewidth=3, markersize=8, 
         color=(0.9, 0.3, 0.3),  # RGB tuple
         label='Expenses', alpha=0.7)

plt.plot(df['Month'], df['Profit'], 
         marker='^', linewidth=3, markersize=10, 
         color='mediumseagreen',  # Named color
         label='Profit', linestyle='--')

plt.xlabel('Month', fontsize=13, fontweight='bold')
plt.ylabel('Amount ($)', fontsize=13, fontweight='bold')
plt.title('Multiple Color Specification Methods', fontsize=15, fontweight='bold', pad=15)
plt.legend(fontsize=11, loc='upper left', framealpha=0.95, shadow=True)
plt.grid(True, alpha=0.25, linestyle=':', color='gray')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Marker Styles and Sizes

Learn about different marker options for data points:

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Marker styles: 'o', 's', '^', 'v', '<', '>', 'D', 'p', '*', 'h', 'H', '+', 'x', 'd', '|', '_'
markers = ['o', 's', '^', 'D']
titles = ['Circle Markers', 'Square Markers', 'Triangle Markers', 'Diamond Markers']

for idx, (ax, marker, title) in enumerate(zip(axes.flat, markers, titles)):
    ax.plot(df['Month'], df['Sales'], 
            marker=marker, 
            markersize=12,
            linewidth=2.5,
            color=f'C{idx}',
            markerfacecolor='white',
            markeredgecolor=f'C{idx}',
            markeredgewidth=2.5,
            label=f'Marker: {marker}')
    
    ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
    ax.set_xlabel('Month', fontsize=11)
    ax.set_ylabel('Sales ($)', fontsize=11)
    ax.legend(fontsize=10)
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 6. Line Styles and Widths

Explore different line styles, widths, and dash patterns:

```python
plt.figure(figsize=(14, 7))

# Line styles: '-' (solid), '--' (dashed), '-.' (dash-dot), ':' (dotted)
# Custom dashes: dashes=[5, 2, 1, 2] means 5pt line, 2pt space, 1pt line, 2pt space

plt.plot(df['Month'], df['Sales'], 
         linestyle='-', linewidth=4, color='#3498DB', 
         alpha=0.7, label='Solid (width=4)')

plt.plot(df['Month'], df['Expenses'], 
         linestyle='--', linewidth=3, color='#E74C3C', 
         alpha=0.8, label='Dashed (width=3)')

plt.plot(df['Month'], df['Profit'], 
         linestyle='-.', linewidth=3, color='#2ECC71', 
         alpha=0.8, label='Dash-dot (width=3)')

# Custom dash pattern
sales_shifted = df['Sales'] * 0.8
plt.plot(df['Month'], sales_shifted, 
         linestyle=(0, (5, 2, 1, 2)), linewidth=2.5, color='#9B59B6',
         alpha=0.8, label='Custom dash pattern')

plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Amount ($)', fontsize=14, fontweight='bold')
plt.title('Line Style Variations', fontsize=16, fontweight='bold', pad=20)
plt.legend(fontsize=12, loc='upper left', framealpha=0.95)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45, fontsize=11)
plt.tight_layout()
plt.show()
```

## 7. Bar Chart Customization

Customize bar colors, edge colors, width, and spacing:

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left plot: Color gradients
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(df)))
axes[0].bar(df['Month'], df['Profit'], 
           color=colors,           # Gradient colors
           edgecolor='black',      # Border color
           linewidth=2,            # Border width
           alpha=0.85,             # Transparency
           width=0.7)              # Bar width (0-1)

axes[0].set_xlabel('Month', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Profit ($)', fontsize=13, fontweight='bold')
axes[0].set_title('Gradient Color Bars', fontsize=14, fontweight='bold', pad=15)
axes[0].grid(axis='y', alpha=0.3, linestyle='--')
axes[0].tick_params(axis='x', rotation=45)

# Right plot: Pattern fills and custom spacing
x_pos = np.arange(len(df['Month']))
axes[1].bar(x_pos, df['Sales'], 
           color='#FF6B6B',
           edgecolor='darkred',
           linewidth=2.5,
           alpha=0.7,
           width=0.6,              # Narrower bars = more spacing
           hatch='//')             # Pattern: '//', '\\\\', '||', '--', '++', 'xx', 'oo', 'OO', '.', '*'

axes[1].set_xlabel('Month', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Sales ($)', fontsize=13, fontweight='bold')
axes[1].set_title('Patterned Bars with Custom Spacing', fontsize=14, fontweight='bold', pad=15)
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(df['Month'], rotation=45)
axes[1].grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```

## 8. Scatter Plot Customization

Customize scatter plots with varying sizes, colors, and transparency:

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left plot: Size and color mapping
sizes = df['Expenses'] / 50  # Scale expenses to point sizes
scatter1 = axes[0].scatter(df['Sales'], df['Profit'], 
                          s=sizes,                    # Variable sizes
                          c=df['Expenses'],           # Color by expenses
                          cmap='plasma',              # Color map: 'viridis', 'plasma', 'coolwarm', 'RdYlGn'
                          alpha=0.6,                  # Transparency
                          edgecolors='black',         # Border color
                          linewidths=2)               # Border width

cbar1 = plt.colorbar(scatter1, ax=axes[0])
cbar1.set_label('Expenses ($)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Sales ($)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Profit ($)', fontsize=13, fontweight='bold')
axes[0].set_title('Variable Size and Color Mapping', fontsize=14, fontweight='bold', pad=15)
axes[0].grid(True, alpha=0.3, linestyle='--')

# Right plot: Different marker shapes and fixed colors
markers_list = ['o', 's', '^', 'D', 'v', '*', 'p', 'h', 'X', 'P', '<', '>']
colors_list = plt.cm.tab20(np.linspace(0, 1, len(df)))

for i in range(len(df)):
    axes[1].scatter(df['Sales'][i], df['Profit'][i], 
                   s=250,                          # Fixed large size
                   marker=markers_list[i],         # Different marker per point
                   color=colors_list[i],
                   alpha=0.7,
                   edgecolors='black',
                   linewidths=1.5,
                   label=df['Month'][i])

axes[1].set_xlabel('Sales ($)', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Profit ($)', fontsize=13, fontweight='bold')
axes[1].set_title('Different Markers per Data Point', fontsize=14, fontweight='bold', pad=15)
axes[1].legend(fontsize=9, ncol=3, loc='upper left', framealpha=0.9)
axes[1].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# Create figure with custom spacing between subplots
fig = plt.figure(figsize=(16, 10))  # Width x Height in inches

# Using gridspec for more control over spacing
gs = fig.add_gridspec(2, 2, 
                     hspace=0.35,     # Horizontal space between plots
                     wspace=0.3,      # Vertical space between plots
                     left=0.08,       # Left margin
                     right=0.95,      # Right margin
                     bottom=0.08,     # Bottom margin
                     top=0.93)        # Top margin

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# Plot 1: Tight spacing example
ax1.plot(df['Month'], df['Sales'], marker='o', linewidth=2.5, color='#E91E63', markersize=8)
ax1.set_title('Plot 1: Line Chart', fontsize=13, fontweight='bold', pad=12)
ax1.set_xlabel('Month', fontsize=11, labelpad=10)  # labelpad controls spacing from axis
ax1.set_ylabel('Sales ($)', fontsize=11, labelpad=10)
ax1.tick_params(axis='x', rotation=45, labelsize=10, pad=5)  # pad controls tick label spacing
ax1.grid(True, alpha=0.3)

# Plot 2: Bar chart
ax2.bar(df['Month'], df['Profit'], color='#4CAF50', edgecolor='darkgreen', linewidth=2, alpha=0.8)
ax2.set_title('Plot 2: Bar Chart', fontsize=13, fontweight='bold', pad=12)
ax2.set_xlabel('Month', fontsize=11, labelpad=10)
ax2.set_ylabel('Profit ($)', fontsize=11, labelpad=10)
ax2.tick_params(axis='x', rotation=45, labelsize=10, pad=5)
ax2.grid(axis='y', alpha=0.3)

# Plot 3: Scatter plot
ax3.scatter(df['Sales'], df['Profit'], s=150, c=df['Expenses'], cmap='coolwarm', 
           edgecolors='black', linewidths=1.5, alpha=0.7)
ax3.set_title('Plot 3: Scatter Plot', fontsize=13, fontweight='bold', pad=12)
ax3.set_xlabel('Sales ($)', fontsize=11, labelpad=10)
ax3.set_ylabel('Profit ($)', fontsize=11, labelpad=10)
ax3.tick_params(labelsize=10, pad=5)
ax3.grid(True, alpha=0.3)

# Plot 4: Multiple lines
ax4.plot(df['Month'], df['Sales'], marker='o', linewidth=2, label='Sales', color='C0')
ax4.plot(df['Month'], df['Expenses'], marker='s', linewidth=2, label='Expenses', color='C1')
ax4.plot(df['Month'], df['Profit'], marker='^', linewidth=2, label='Profit', color='C2')
ax4.set_title('Plot 4: Multi-line Chart', fontsize=13, fontweight='bold', pad=12)
ax4.set_xlabel('Month', fontsize=11, labelpad=10)
ax4.set_ylabel('Amount ($)', fontsize=11, labelpad=10)
ax4.legend(fontsize=10, loc='upper left', framealpha=0.9)
ax4.tick_params(axis='x', rotation=45, labelsize=10, pad=5)
ax4.grid(True, alpha=0.3)

fig.suptitle('Custom Spacing and Layout Control', fontsize=16, fontweight='bold', y=0.98)
plt.show()
```

 Introduction to Linear Regression lines and Residual Plots:

```python
plt.figure(figsize=(10, 6))

# Plot the actual data: Profit vs Expenses
plt.scatter(df['Profit'], df['Expenses'], 
       marker='s', s=100, 
       color='#E74C3C',
       edgecolors='darkred',
       linewidths=2,
       alpha=0.85,
       label='Actual Data',
       zorder=2)

# Calculate line of best fit using z values
# Profit is x, Expenses is y
z = np.polyfit(df['Profit'], df['Expenses'], 1)  # Returns [slope, intercept]
p = np.poly1d(z)  # Creates polynomial function

# Plot the regression line
plt.plot(df['Profit'], p(df['Profit']), 
         linewidth=2.5, 
         color='darkblue', 
         linestyle='-',
         label=f'Best Fit Line: y = {z[0]:.2f}x + {z[1]:.2f}',
         zorder=1)

plt.xlabel('Profit ($)', fontsize=12, fontweight='bold')
plt.ylabel('Expenses ($)', fontsize=12, fontweight='bold')
plt.title('Profit vs Expenses with Line of Best Fit', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Now for the Residual Plot:

```python
plt.figure(figsize=(12, 5))

# Calculate regression for Profit vs Expenses
z = np.polyfit(df['Profit'], df['Expenses'], 1)
p = np.poly1d(z)

# Calculate predicted values
predicted_expenses = p(df['Profit'])

# Calculate residuals (actual - predicted)
residuals = df['Expenses'] - predicted_expenses

# Create residual plot
plt.scatter(df['Month'], residuals, 
           marker='o', s=100, 
           color="#144E58",
           edgecolors='purple',
           linewidths=2,
           alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Residual Line')
plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Residuals ($)', fontsize=12, fontweight='bold')
plt.title('Residual Plot: Expenses vs Profit', fontsize=13, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()

```

## Introduction to Plotly: Interactive Visualizations

So far we've been using Matplotlib, which creates static images. **Plotly** is a powerful library for creating **interactive** visualizations that allow you to:
- Hover over data points to see exact values
- Zoom in and out
- Pan across the plot
- Toggle series on/off

Plotly is particularly useful for exploratory data analysis and presentations where you want your audience to interact with the data.

```python
#Run if you need to install plotly
#!pip install plotly
```

```python
# First, let's install and import plotly
import plotly.express as px
import plotly.graph_objects as go
```

### 1. Basic Line Plot with Plotly Express

Plotly Express (px) is the high-level API that makes creating common plots very easy. Let's recreate a simple line plot.

```python
# Using our existing dataframe
fig = px.line(df, x='Month', y='Sales', 
              title='Monthly Sales (Interactive)',
              labels={'Sales': 'Sales ($)', 'Month': 'Month'})

# Customize the appearance
fig.update_traces(line_color='#2E86AB', line_width=3)
fig.update_layout(
    hovermode='x unified',  # Shows all values for the x-value you're hovering over
    plot_bgcolor='white',
    font=dict(size=12)
)

fig.show()

# Try hovering over the line, zooming in, and clicking the camera icon to save an image!
```

### 2. Multiple Lines on One Plot

Just like Matplotlib, we can plot multiple series. With Plotly, each line is automatically interactive!

```python
fig = go.Figure()

# Add multiple traces (lines)
fig.add_trace(go.Scatter(x=df['Month'], y=df['Sales'], 
                         mode='lines+markers',
                         name='Sales',
                         line=dict(color='#2E86AB', width=3),
                         marker=dict(size=8)))

fig.add_trace(go.Scatter(x=df['Month'], y=df['Expenses'], 
                         mode='lines+markers',
                         name='Expenses',
                         line=dict(color='#A23B72', width=3),
                         marker=dict(size=8)))

fig.add_trace(go.Scatter(x=df['Month'], y=df['Profit'], 
                         mode='lines+markers',
                         name='Profit',
                         line=dict(color='#F18F01', width=3),
                         marker=dict(size=8)))

# Customize layout
fig.update_layout(
    title='Sales, Expenses, and Profit Over Time',
    xaxis_title='Month',
    yaxis_title='Amount ($)',
    hovermode='x unified',
    legend=dict(x=0, y=1, bgcolor='rgba(255,255,255,0.8)'),
    plot_bgcolor='white'
)

fig.show()

# Click on legend items to hide/show lines!
```

### 3. Interactive Scatter Plot

Scatter plots are particularly useful in Plotly because you can hover to see exact values and identify specific points.

```python
# Create an interactive scatter plot with color and size encoding
fig = px.scatter(df, x='Sales', y='Profit', 
                 size='Expenses',           # Bubble size based on Expenses
                 color='Expenses',          # Color based on Expenses
                 hover_data=['Month'],      # Show Month when hovering
                 title='Sales vs Profit (Size and Color = Expenses)',
                 color_continuous_scale='Plasma')  # Color scheme

# Customize
fig.update_traces(marker=dict(line=dict(width=2, color='DarkSlateGrey')))
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(title='Sales ($)', gridcolor='lightgray'),
    yaxis=dict(title='Profit ($)', gridcolor='lightgray')
)

fig.show()

# Hover over points to see all details including the month!
```

# Generating an Regression Line in Plotly:

```python
# Calculate regression line first
z = np.polyfit(df['Sales'], df['Profit'], 1)  # Returns [slope, intercept]
p = np.poly1d(z)  # Creates a polynomial function

# Add predicted values to the dataframe
df['Predicted_Profit'] = p(df['Sales'])

# Create scatter plot with predicted values in hover data
fig = px.scatter(df, x='Sales', y='Profit', 
                 size='Expenses',
                 color='Expenses',
                 hover_data={
                     'Month': True,           # Show Month
                     'Predicted_Profit': ':.2f',  # Show predicted value with 2 decimals
                     'Expenses': ':.0f'       # Show Expenses with no decimals
                 },
                 title='Sales vs Profit with Regression Line (Size and Color = Expenses)',
                 color_continuous_scale='Plasma')

# Customize scatter points
fig.update_traces(marker=dict(line=dict(width=2, color='DarkSlateGrey')))
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(title='Sales ($)', gridcolor='lightgray'),
    yaxis=dict(title='Profit ($)', gridcolor='lightgray')
)

# Add regression line as a separate trace
fig.add_trace(go.Scatter(
    x=df['Sales'], 
    y=p(df['Sales']),
    mode='lines',
    name=f'Regression Line<br>y = {z[0]:.3f}x + {z[1]:.2f}',
    line=dict(color='red', width=3)
))
                        
fig.show()
```

### 4. Bar Charts with Plotly

Bar charts are cleaner and more interactive with Plotly. You can hover to see exact values and easily compare bars.

```python
# Grouped bar chart
fig = go.Figure()

fig.add_trace(go.Bar(x=df['Month'], y=df['Sales'], 
                     name='Sales',
                     marker_color='#2E86AB'))

fig.add_trace(go.Bar(x=df['Month'], y=df['Expenses'], 
                     name='Expenses',
                     marker_color='#A23B72'))

fig.add_trace(go.Bar(x=df['Month'], y=df['Profit'], 
                     name='Profit',
                     marker_color='#F18F01'))

fig.update_layout(
    title='Monthly Financial Data Comparison',
    xaxis_title='Month',
    yaxis_title='Amount ($)',
    barmode='group',  # Try changing to 'stack' for stacked bars!
    plot_bgcolor='white',
    hovermode='x unified'
)

fig.show()
```

### 5. Key Differences: Matplotlib vs Plotly

**Matplotlib:**
- Creates static images (PNG, PDF)
- Great for publication-quality figures
- More control over fine details
- Works offline without browser

**Plotly:**
- Creates interactive HTML visualizations
- Better for exploratory analysis and presentations
- Built-in interactivity (zoom, pan, hover)
- Easier to create complex visualizations quickly
- Best viewed in browser/notebook

**When to use which?**
- Use **Matplotlib** for papers, reports, and when you need pixel-perfect control
- Use **Plotly** for dashboards, presentations, and when exploring data interactively
